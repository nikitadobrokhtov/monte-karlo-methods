```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from mpl_toolkits.mplot3d.art3d import Line3DCollection
```

```python
E_pot_30 = np.load("./T_30/E_pot.npy")
E_kin_30 = np.load("./T_30/E_kin.npy")

E_pot_300 = np.load("./T_300/E_pot.npy")
E_kin_300 = np.load("./T_300/E_kin.npy")
```

```python
fig = plt.figure(figsize=(12, 9))

plt.plot(E_pot_30, label="$E_{pot}, T=30K$")
plt.plot(E_kin_30, label="$E_{kin}, T=30K$")

plt.plot(E_pot_300, label="$E_{pot}, T=300K$")
plt.plot(E_kin_300, label="$E_{kin}, T=300K$")


plt.xlabel("Step", fontsize=14)
plt.ylabel("Energy", fontsize=14)

plt.title("Energy stabilization",fontsize=20, color='red')
plt.legend()

plt.show()
```

```python
accepted_30 = np.load("./T_30/acc.npy")
non_accepted_30 = np.load("./T_30/non_ac.npy")

accepted_300 = np.load("./T_300/acc.npy")
non_accepted_300 = np.load("./T_300/non_ac.npy")

print("T = 30K; Accepted: ", accepted_30, ", Unaccepted: ", non_accepted_30)
print("T = 300K; Accepted: ", accepted_300, ", Unaccepted: ", non_accepted_300)
```

```python
list_of_atoms_30 = np.load("./T_30/list_of_atoms.npy")
list_of_atoms_300 = np.load("./T_300/list_of_atoms.npy")
```

```python
def plot_lattice(lattice, title):
    
    def list_zip(coords_of_atom):
        m = np.concatenate((coords_of_atom, [coords_of_atom[0]]), axis=0).T
        return [list(zip(m[0], m[1], m[2]))]

    fig = plt.figure(figsize=(12, 9))
    ax = fig.add_subplot(111, projection='3d')
    
    
    for coord in lattice:
        ax.add_collection3d(Line3DCollection(list_zip(coord)))
        for point in coord:
            ax.scatter(point[0], point[1], point[2], color='blue', s=7)
        
    
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.title(title, fontsize=20, color='red')
    
    plt.show()   
```

```python
plot_lattice(list_of_atoms_30[0], "T=30K, end state")
```

```python
plot_lattice(list_of_atoms_300[0], "T=300K, end state")
```
