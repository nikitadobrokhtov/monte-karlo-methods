```python
#Import all appropriate libraries
import numpy as np
import matplotlib.pyplot as plt

#Load our calculated data
p = np.load('./lattice_prob.npy')
P_5 = np.load('./inf_clust_prob_5.npy')
P_10 = np.load('./inf_clust_prob_10.npy')
P_15 = np.load('./inf_clust_prob_15.npy')
P_30 = np.load('./inf_clust_prob_30.npy')
P_100= np.load('./inf_clust_prob_100.npy')

p_30_thorough = np.load('./lattice_prob_10000.npy')
P_30_thorough = np.load('./inf_clust_prob_30_thorough_10000.npy')
```

# Построим графики зависимостей в промежутке вероятностей $p\in[0.3,0.8]$. Также построим прямую $P=\frac{1}{2}$

```python
fig = plt.figure(figsize=(12, 9))

plt.plot(p, P_5, label='P, $N\\times N = 5\\times 5$')
plt.plot(p, P_10, label='P, $N\\times N = 10\\times 10$')
plt.plot(p, P_15, label='P, $N\\times N = 15\\times 15$')
plt.plot(p, P_30, label='P, $N\\times N = 30\\times 30$')
plt.plot(p, P_100, label='P, $N\\times N = 100\\times 100$')


plt.plot(p, np.ones(len(p)) * 1/2, color='deeppink', linestyle='--', label = 'P = $\\frac{1}{2}$')

plt.xlabel("p, probability of node existence", fontsize=14)
plt.ylabel("P, probability of infinite cluster", fontsize=14)
plt.title("P(p) dependency", fontsize=20, color='red')

plt.legend()
plt.savefig("percolation.png")
plt.show()
```

# Как видим, с увеличением размера решётки (номер $N$), график зависимости $P(p)$ становится более "крутым". Это согласуется с теорией: в пределе $\lim\limits_{N\rightarrow+\infty}$ график выродится просто в ступеньку, где скачок будет происходить при пороге перколяции $p$. 

# Попробуем оценить эту величину $p$, при которой $P(p)=\frac{1}{2}$. Для этого построем графики более подробно в той области, где они пересекают прямую $P(p)=\frac{1}{2}$.

```python
fig = plt.figure(figsize=(12, 9))

plt.plot(p, P_5, label='P, $N\\times N = 5\\times 5$')
plt.plot(p, P_10, label='P, $N\\times N = 10\\times 10$')
plt.plot(p, P_15, label='P, $N\\times N = 15\\times 15$')
plt.plot(p, P_30, label='P, $N\\times N = 30\\times 30$')
plt.plot(p, P_100, label='P, $N\\times N = 100\\times 100$')


plt.plot(p, np.ones(len(p)) * 1/2, color='deeppink', linestyle='--', label = 'P = $\\frac{1}{2}$')

plt.xlabel("p, probability of node existence", fontsize=14)
plt.ylabel("P, probability of infinite cluster", fontsize=14)
plt.title("P(p) dependance", fontsize=20, color='red')

plt.xlim(0.58, 0.6)

plt.legend()

plt.show()
```

# Так как значение $p$ будет тем точнее, чем выше номер $N$, рассмотрим случай $N=100$, а также более тщательно рассчитанный случай $N=30$ на промежутке $p\in[0.58,0.6]$ (10000 раз на каждое значение вероятности $P$ + более мелкий шаг по вероятности $p$).

```python
fig = plt.figure(figsize=(12, 9))

plt.plot(p, P_30, label='P, $N\\times N = 30\\times 30$; 1000 expers')
plt.plot(p, P_100, label='P, $N\\times N = 100\\times 100$; 1000 expers')
plt.plot(p_30_thorough, P_30_thorough, label='P, $N\\times N = 30\\times 30$; 10000 expers')

plt.plot(p, np.ones(len(p)) * 1/2, color='deeppink', linestyle='--', label = 'P = $\\frac{1}{2}$')

plt.xlabel("p, probability of node existence", fontsize=14)
plt.ylabel("P, probability of infinite cluster", fontsize=14)
plt.title("P(p) dependance", fontsize=20, color='red')

plt.xlim(0.58, 0.6)

plt.legend()

plt.show()
```

# Распечатаем некоторые значения вблизи $P(p)\approx\frac{1}{2}$

```python
print('$N=100$')
for i in range(290, 300):
    print("i = " + str(i) + "; P = " + str(P_100[i]) + "; p = " + str(p[i]))
```

```python
print('N=30, thorough')
for i in range(110, 125):
    print("i = " + str(i) + "; P = " + str(P_30_thorough[i]) + "; p = " + str(p_30_thorough[i]))
```

# Аппроксимируем прямой эти значения и найдём точку пересечения этих прямых с прямой $P(p)=\frac{1}{2}$. Получим следующие значения:

```python
print('N=100')
coefs_100 = np.polyfit(p[290:300], P_100[290:300], 1)
p_final_100 = (1/2 - coefs_100[1]) / coefs_100[0]
print(p_final_100)
```

```python
print('N=30, thorough')
coefs_30_thorough = np.polyfit(p_30_thorough[110:125], P_30_thorough[110:125], 1)
p_final_30_thorough = (1/2 - coefs_30_thorough[1]) / coefs_30_thorough[0]
print(p_final_30_thorough)
```

```python
fig = plt.figure(figsize=(12, 9))

plt.plot(p, p * coefs_100[0] + coefs_100[1], 
         label='P, $N\\times N = 100\\times 100$; 1000 expers; approximation')
plt.plot(p_30_thorough, p_30_thorough * coefs_30_thorough[0] + coefs_30_thorough[1], 
         label='P, $N\\times N = 30\\times 30$; 10000 expers; approximation')

plt.axvline(x=0.5925510467821139, label='Value for $N\\times N = 100\\times 100$', color='green')
plt.axvline(x=0.5914811568968853, label='Value for $N\\times N = 30\\times 30$', color='purple')
plt.axvline(x=0.5927, label='Value from Wikipedia', color='black', linewidth=4.0, linestyle='--')


plt.plot(p, np.ones(len(p)) * 1/2, color='deeppink', linestyle='--', label = 'P = $\\frac{1}{2}$')

plt.xlabel("p, probability of node existence", fontsize=14)
plt.ylabel("P, probability of infinite cluster", fontsize=14)
plt.title("P(p) dependance", fontsize=20, color='red')

plt.xlim(0.58, 0.6)

plt.legend()

plt.show()
```

# Видим, что несмотря на то, что для $N=30$ "траектория" рассчитывалась более подробно, для $N=100$ значение вероятности вышло более точным. Тем не менее, две первые значащие цифры сошлись. Данный эксперимент сошёлся с результатами, полученными в статьях; проверено стремление системы к более резкому переходу при увеличении её размера (с теорией сошлось).
