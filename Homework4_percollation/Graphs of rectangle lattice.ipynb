```python
#Import all appropriate libraries
import numpy as np
import matplotlib.pyplot as plt

#Load our calculated data
p = np.load('./lattice_prob_5_times_10.npy')
P_5_10 = np.load('./inf_clust_prob_5_times_10.npy')
P_10_20 = np.load('./inf_clust_prob_10_times_20.npy')
P_50_100 = np.load('./inf_clust_prob_50_times_100.npy')


```

```python
fig = plt.figure(figsize=(12, 9))

plt.plot(p, P_5_10, label='P, $N\\times M = 5\\times 10$')
plt.plot(p, P_10_20, label='P, $N\\times M = 10\\times 20$')
plt.plot(p, P_50_100, label='P, $N\\times M = 50\\times 100$')




plt.plot(p, np.ones(len(p)) * 1/2, color='deeppink', linestyle='--', label = 'P = $\\frac{1}{2}$')

plt.xlabel("p, probability of node existence", fontsize=14)
plt.ylabel("P, probability of infinite cluster", fontsize=14)
plt.title("P(p) dependance", fontsize=20, color='red')

plt.legend()

plt.show()
```
